```python
altsorgu=veritabani['speaker']=="Lily Yeh" #speaker "Lily Yeh" olan sorgu.
veritabani[altsorgu]
```

```python
altsorgu1=veritabani['likes']>50000 #Beğenisi 50000'den fazla olan sorgular.
veritabani[altsorgu1]
```

```python
altsorgu2=veritabani['views']>10000000 #views 10 000 000'den fazla olan sorgular.
veritabani[altsorgu2]
```

```python
altsorgu3=veritabani['title']=="Do schools kill creativity?" #title "Do schools kill creativity?" olan sorgular.
veritabani[altsorgu3]
```

```python
veritabani.sort_values("recorded_date") #recorded_date'e göre sıralama.
```

```python
altsorgu5=veritabani['event']=="TED2006" #events "tes2006" olan sorgular.
veritabani[altsorgu5]
```

```python
altsorgu4=veritabani["published_date"].str.contains('2006', case=False, regex=True)  #published_date'i 2006 olan olan sorgular.
veritabani[altsorgu4]
```

```python
altsorgu6=veritabani["title"].str.contains('wish', case=False, regex=True)         #title'da "wish" geçen sorgular.
veritabani[altsorgu6]
```

```python
veritabani.plot() #Başlıklara göre dağılımı.
plt.show()
```

```python
veritabani.plot(kind = 'scatter', x = 'talk_id', y = 'likes') #like'a göre nokta dağılımı.
plt.show() 
```

```python
veritabani.plot(kind = 'scatter', x = 'talk_id', y = 'duration') #duration'a göre nokta dağılımı.
plt.show()
```

```python
veritabani.plot(kind = 'hist') #Histograma göre dağılımı 
plt.show()
```

```python
altsorgu6=veritabani['event']=="TED2006" #tes2006 event katılanların like dağlımı
veritabani[altsorgu6].plot(kind = 'bar', x = 'talk_id', y = 'likes') 
plt.show()
```

```python
altsorgu7=veritabani.tail(20)  #Son 20 verinin like dağılımı
altsorgu7.plot(kind = 'bar', x = 'talk_id', y = 'likes') 
plt.show()
```
